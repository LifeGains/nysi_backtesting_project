# EMA Calc

```python
# EMA Function
def exponential_moving_average(prices, period=8):
    alpha = 2 / (period + 1)
    ema_values = [np.nan] * (period - 1)  # Fill the first (period - 1) values with 'NA'

    ema_values.append(sum(prices[:period]) / period)

    for i in range(period, len(prices)):
        ema_today = (prices[i] * alpha) + (ema_values[-1] * (1 - alpha))
        ema_values.append(ema_today)

    return ema_values
```

```python
df['Close'] = df['Close'].astype(float)
df
```

```python
df['EMA8'] = exponential_moving_average(df['Close'].tolist(), period=8)
df
```

# Create new column called "NYSIPosition", which returns the values "Above" or "Below" based on if the value of "Close" is Above or Below the value in "EMA8".

```python
df['NYSIPosition'] = np.where(df['Close'] >= df['EMA8'], 'Above', 'Below')
df
```

# Create new column called "Signal". Whenever the NYSIPosition column flips from "Above" to "Below", return "SignalDown". Whenever the NYSIPosition column flips from "Below" to "Above", return "SignalUp". Else, return "NoSignal".

```python
df['PrevNYSIPosition'] = df['NYSIPosition'].shift(1)
conditions = [
    (df['NYSIPosition'] == 'Above') & (df['PrevNYSIPosition'] == 'Below'),
    (df['NYSIPosition'] == 'Below') & (df['PrevNYSIPosition'] == 'Above')
]

choices = ['SignalUp', 'SignalDown']

df['Signal'] = np.select(conditions, choices, default='NoSignal')
df
```

# Add SPX Change

- In the following dataframe, if the df['Signal'] == 'SignalUp', go down the dataframe to find the next row where df['Signal'] == 'SignalDown', ignoring any 'NoSignal' entries. Use that row to find the percentage change of (SignalDown.spy_Close/SignalUp.spy_Close - 1).
Return the results in a new column called 'spy_Pct_Chg'. Fill all the rows where df['Signal'] == 'NoSignal' as 'NA'.

If the df['Signal'] == 'SignalUp', go down the dataframe to find the next row where df['Signal'] == 'SignalDown', ignoring any 'NoSignal' entries. Use that row to find the percentage change of (SignalDown.spy_Close/SignalUp.spy_Close - 1).

```python
# Verify that signalup and signaldown flipflop.
df['Signal'].value_counts()
```

```python
df['spy_Close'] = df['spy_Close'].astype(float)
```

```python
# Initialize the new column with 'NA' values
df['spy_Pct_Chg'] = 'NA'

# Find the indexes where the 'SignalUp' and 'SignalDown' are located
signal_up_indexes = df[df['Signal'] == 'SignalUp'].index
signal_down_indexes = df[df['Signal'] == 'SignalDown'].index

# Iterate over the 'SignalUp' indexes
for up_idx in signal_up_indexes:
    # Find the first 'SignalDown' index that comes after the current 'SignalUp' index
    down_idx = signal_down_indexes[signal_down_indexes > up_idx].min()

    # Calculate the percentage change and set it for the 'SignalUp' rows
    if not np.isnan(down_idx):
        pct_chg = (df.loc[down_idx, 'spy_Close'] / df.loc[up_idx, 'spy_Close'] - 1) * 100
        df.loc[up_idx, 'spy_Pct_Chg'] = pct_chg

# Iterate over the 'SignalDown' indexes
for down_idx in signal_down_indexes:
    # Find the first 'SignalUp' index that comes after the current 'SignalDown' index
    up_idx = signal_up_indexes[signal_up_indexes > down_idx].min()

    # Calculate the percentage change and set it for the 'SignalDown' rows
    if not np.isnan(up_idx):
        pct_chg = -(df.loc[up_idx, 'spy_Close'] / df.loc[down_idx, 'spy_Close'] - 1) * 100
        df.loc[down_idx, 'spy_Pct_Chg'] = pct_chg

df[15:45]
```

```python
# Add 1 to each percentage change
df_signals['spy_Pct_Chg_1'] = df_signals['spy_Pct_Chg']/100 + 1

# Set the index to the timestamps
df_signals.index.name = 'Date'
df_signals.reset_index(inplace=True)

# Group the dataframe by year
yearly_groups = df_signals.resample('Y', on='Date')

# Initialize an empty DataFrame to store the results
df_yearly = pd.DataFrame(columns=['Year', 'Total_Pct_Chg'])

# Loop through the groups and calculate the cumulative product for each group
for year, group in yearly_groups:
    if not group.empty:
        cumprod = group['spy_Pct_Chg_1'].cumprod().iloc[-1]
        df_yearly = df_yearly.append({'Year': year, 'Total_Pct_Chg': (cumprod - 1) * 100}, ignore_index=True)

# Convert the Year column to datetime and then to integer
df_yearly['Year'] = df_yearly['Year'].dt.year
```
